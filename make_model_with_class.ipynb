```
import torch
import torch.nn as nn
import torch.nn.functional as F
```

클래스(class) 형태의 모델은 nn.Module 을 상속받습니다.  
그리고 __init__()에서 모델의 구조와 동적을 정의하는 생성자를 정의합니다.  
이는 파이썬에서 객체가 갖는 속성값을 초기화하는 역할로,  
객체가 생성될 때 자동으호 호출됩니다.  
  
super() 함수를 부르면   
여기서 만든 클래스는 nn.Module 클래스의 속성들을 가지고 초기화 됩니다.     

foward() 함수는 모델이 학습데이터를 입력받아서   
forward 연산을 진행시키는 함수입니다.   
이 forward() 함수는 model 객체를 데이터와 함께 호출하면 자동으로 실행이됩니다.  
  
예를 들어 model이란 이름의 객체를 생성 후,   
model(입력 데이터)와 같은 형식으로 객체를 호출하면  
자동으로 forward 연산이 수행됩니다.

- H(x) 식에 입력 x로부터 예측된 y를 얻는 것을 forward 연산이라고 합니다.

```
#-------------------
model = nn.Linear(1,1)

#---위에 것을 class로 구현하면 아래와 같음---#

class LinearRegressionModel(nn.Module):

    def __init__(self):
        super().__init__()
        #단순 선형 회귀이므로 input_dim = 1, output_dim = 1
        self.linear = nn.Linear(1,1) 

    def forward(self, x):
        return self.linear(x)
        #nn.Linear(1,1)(x)
```

```
model = LinearRegressionModel()
```

```
class MultivariateLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        #다중선형회귀이므로 input_dim = 3, out_dim = 1
        self.linear = nn.Linear(3,1) 

    def forward(self,x):
        return self.linear(x)
        #nn.Linear(1,1)(x)
```

```
model = MultivariateLinearRegressionModel()
```

# 2. 단순 선형 회귀 클래스로 구현하기

```
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```
torch.manual_seed(1)
```

```
# 데이터
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])

print(x_train.shape, y_train.shape)
```

```
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1,1)

    def forward(self,x):
        return self.linear(x)
        #nn.Linear(1,1)(x)
```

```
model = LinearRegressionModel()
```

```
print(type(model.parameters()))


optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```
nb_epochs = 2000

for epoch in range(nb_epochs + 1):

    prediction = model(x_train)
    cost = F.mse_loss(prediction, y_train)
    # print(type(cost))
    # ---> torch.Tensor

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(epoch, nb_epochs, cost.item())
```

# 3. 다중선형회귀 

```
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```
torch.manual_seed(1)
```

```
# 데이터
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

print(x_train.shape, y_train.shape)
```

```
class MultivariateLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()

        #사이즈 잡아줘야함
        self.linear = nn.Linear(3, 1) # 다중 선형 회귀이므로 input_dim=3, output_dim=1.

    def forward(self, x):
        return self.linear(x)
```

```
model = MultivariateLinearRegressionModel()
```

```
optimizer = torch.optim.SGD(model.parameters(), lr = 1e-5)
```

```
nb_epochs = 2000
for epoch in range(nb_epochs + 1):
    
    #h(x)
    prediction = model(x_train)
    #model(x_train) = model.forward(x_train)과 동일하다 

    #cost
    cost = F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    #미분해서 gradient 계산
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
    # 100번마다 로그 출력
      print('{}\nEpoch {:4d}/{} Cost: {:.6f},  \n{}'.format(
          '-'*70,epoch, nb_epochs, cost.item(), list(model.parameters())
      ))
```
